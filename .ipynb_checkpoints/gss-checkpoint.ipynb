```python
%matplotlib inline
import numpy as np
import sympy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# example equation
def f_1(x1,x2):
    return -x2-2*x1*x2-x1**3-x1**2*x2-x1*x2**2
def f_2(x1,x2):
    return x1-x1**2-3*x1**3-x1**2*x2-x2**3
def f((x1,x2),t):
    return np.array([f_1(x1,x2),f_2(x1,x2)])
```

```python
x1 = np.linspace(-1,1)
x2 = np.linspace(-1,1)
xx1,xx2 = np.meshgrid(x1,x2)
# nullclines
plt.contour(xx1,xx2,f_1(xx1,xx2),[0])
plt.contour(xx1,xx2,f_2(xx1,xx2),[0])
# flow
plt.streamplot(xx1,xx2,f_1(xx1,xx2),f_2(xx1,xx2),color='.5')
plt.savefig('./fig/example.pdf')
```

```python
# lie: compute 2(j+k) by 2(j+1) matrix representation of Lie bracket [.,g] on polynomial vector field
# j: degree of vector field
# g: vector of length 2k+2 that represents degree k generator 
# n: dimension of vector field
def lie(j,g,n=2):
    x_ = sp.symbols('x1 x2')
    # produce canonical basis for jth degree n-dim vector fields
    def vf_basis(j,n):
        bas_ = [sp.Matrix([[x_[0]**(j-i)*x_[1]**i],[0.]]) for i in range(j+1)]
        bas_ += [sp.Matrix([[0.],[x_[0]**(j-i)*x_[1]**i]]) for i in range(j+1)]
        return bas_
    # degree of generator g
    k = len(g)/2-1
    # symbolic coefficients of generator g
    g_ = [sp.symbols('g%d' % (i+1)) for i in range(2*k+2)]
    # symbolic vector field version of generator g
    G_ = sp.Matrix([[sum([g_[i]*x_[0]**(k-i)*x_[1]**i for i in range(k+1)])],
                    [sum([g_[i+k+1]*x_[0]**(k-i)*x_[1]**i for i in range(k+1)])]])
    # symbolic derivative of generator g
    DG_ = sp.Matrix([[sp.diff(G_[0],x_[0]),sp.diff(G_[0],x_[1])],
                     [sp.diff(G_[1],x_[0]),sp.diff(G_[1],x_[1])]])
    # canonical basis for jth degree 2-dim vector fields
    bas_ = vf_basis(j,n)
    # derivatives of canonical basis
    Dbas_ = [sp.Matrix([[sp.diff(bas_[i][0],x_[0]),sp.diff(bas_[i][0],x_[1])],
            [sp.diff(bas_[i][1],x_[0]),sp.diff(bas_[i][1],x_[1])]]) for i in range(2*j+2)]
    # symbolic Lie transformation generated by g
    L = sp.zeros(2*(j+k),2*j+2)
    for i in range(2*j+2):
        # lie bracket of g with canonical basis element
        bracket = sp.expand(Dbas_[i]*G_-DG_*bas_[i])
        # determine column of L as coefficients of polynomial terms
        L[:j+k,i] = [bracket[0].coeff(x_[0]**(j+k-1-l)*x_[1]**l) for l in range(j+k)]
        L[j+k:,i] = [bracket[1].coeff(x_[0]**(j+k-1-l)*x_[1]**l) for l in range(j+k)]
    # substitute numerical values and return as float matrix
    for i in range(2*k+2):
        L = L.subs(g_[i],g[i])
    return np.array(L).astype(float)
```

```python
f1 = [0,-1,1,0]
L12 = lie(2, g=f1)
# representation of L_{f_1} on 2nd degree 2-dim polynomial vector fields
print L12
```

```python
# (j=2) calculate modified degree 2 term
g2 = [1,0,0,0,0,0]
f2 = np.array([0,-2,0,1,0,0])
f3 = np.array([-1,-1,-1,0,-3,-1,0,-1])
L21 = lie(1, g=g2)
L22 = lie(2, g=g2)
print f3 + np.dot(L22, f2) + .5 * np.dot(L22, np.dot(L21,f1))
```

```python
# normal form truncated at degree 3
def h_1(x1,x2):
    return -x2-x1**3-x1*x2**2
def h_2(x1,x2):
    return x1-x1**2*x2-x2**3
def h((x1,x2),t):
    return np.array([h_1(x1,x2),h_2(x1,x2)])
```

```python
# nullclines
plt.contour(xx1,xx2,h_1(xx1,xx2),[0])
plt.contour(xx1,xx2,h_2(xx1,xx2),[0])
# flow
plt.streamplot(xx1,xx2,h_1(xx1,xx2),h_2(xx1,xx2),color='.5')
plt.savefig('./fig/normal.pdf')
```

```python
# linearization of f
def fp_1(x1,x2):
    return -x2
def fp_2(x1,x2):
    return x1
def fp((x1,x2),t):
    return np.array([fp_1(x1,x2),fp_2(x1,x2)])
# the 'expansion' equivariant v_1
def v_1(x1,x2):
    return x1
def v_2(x1,x2):
    return x2
def v((x1,x2),t):
    return np.array([v_1(x1,x2),v_2(x1,x2)])
```

```python
y0 = [1./np.sqrt(2),1./np.sqrt(2)]
t = np.linspace(0,1)
# flow along fp then v
sol = odeint(fp,y0,t)
plt.plot(sol[:,0],sol[:,1],'r-')
sol = odeint(v,sol[-1],t)
plt.plot(sol[:,0],sol[:,1],'r-')
# flow along v then fp
sol = odeint(v,y0,t)
plt.plot(sol[:,0],sol[:,1],'b-')
sol = odeint(fp,sol[-1],t)
plt.plot(sol[:,0],sol[:,1],'b-')
plt.savefig('./fig/equivariance.pdf')
```

```python
# trajectories
sol = odeint(h,[1,0],t)
plt.plot(sol[:,0],sol[:,1],'r-')
sol = odeint(h,[0,1],t)
plt.plot(sol[:,0],sol[:,1],'b-')
# leaves
r = np.linspace(0,2*np.pi)
mag = np.sqrt(sum(sol[-1]**2))
plt.plot(np.cos(r),np.sin(r),color='.5')
plt.plot(mag*np.cos(r),mag*np.sin(r),color='.5')
plt.xlim(-1,1); plt.ylim(-1,1)
plt.savefig('./fig/foliation.pdf')
```
